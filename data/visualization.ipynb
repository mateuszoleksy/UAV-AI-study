Import required libraries

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from configuration import *
```

Load proccesed data:

```python
dataset = pd.read_csv(configuration.get_filepath_to_save())
```

Let's jump to the visualizations, first begin with heatmap:

```python
# 1. Calculate the correlation matrix
corr_matrix = dataset.corr()

# 2. Plot the heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
```

Histogram and scatter:

```python
sns.pairplot(dataset)
```

Histogram for every column:

```python
for column in dataset.select_dtypes(include=['number']).columns:
    plt.figure(figsize=(8, 4))      # Create a new "canvas" for each plot
    sns.histplot(dataset[column], kde=True) # kde=True adds a smooth trend line
    plt.title(f'Distribution of {column}')
    plt.show()                      # Show the plot and clear the canvas for the next one
```

Boxplot:

```python
for column in dataset.select_dtypes(include=['number']).columns:
    plt.figure(figsize=(8, 2)) # Boxplots are often better "wide"
    sns.boxplot(x=dataset[column])
    plt.title(f'Outliers in {column}')
    plt.show()
```

Desribe (max, min, deviation):

```python
dataset.describe()
```

Violinplot:

```python
sns.violinplot(dataset)
```
